# HeatMap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.linspace(1, 10, 20).reshape((4, 5))
data
```

```python
# simple heatmap
sns.heatmap(data)
plt.show()
```

```python
# loading data from dataframe

data = sns.load_dataset("anagrams")
data
```

```python
# you cannot create heatmaps for categorical data so drop it.
data.drop(columns=['attnr'], inplace=True)
data
```

```python
# see the heatmap for the remaining data

data = data.head(10) #take values from top 10 data (vrna boht bhasad mach rhi thi)
sns.heatmap(data)

plt.show()
```

```python
# vmin and vmax,cmap and annotations

data = data.head(10) 
sns.heatmap(data, vmin=0, vmax=20, cmap="PuOr", annot=True)

plt.show()
```

```python
# adding custom annotations

# defining 2d data
var = [[1, 2, 3, 4], [5, 6, 7, 8]]

# custom data
cusData = [['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h']]
sns.heatmap(var, annot=cusData, fmt='s') # format(fmt='s') meaning, format of annotation is a string value.
plt.show()
```

```python
# adding more customizations

# defining 2d data
var = [[1, 2, 3, 4], [5, 6, 7, 8]]

design = {"fontsize":20, "color":"g"}
# custom data
cusData = [['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h']]
sns.heatmap(var, annot=cusData, fmt='s', annot_kws=design) 
plt.show()
```

```python
# adding linewidth, linecolor, cbar, xticklabels, yticklabels and set_function

# defining 2d data
var = [[1, 2, 3, 4], [5, 6, 7, 8]]

design = {"fontsize":20, "color":"g"}
# custom data
cusData = [['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h']]

#cbar -> show or remove color bar
#xticklabels / yticklabels=show or hide x ticks or y ticks
v = sns.heatmap(var, annot=cusData, fmt='s', annot_kws=design, linewidth=10, linecolor='y', cbar=False,
            xticklabels=False, yticklabels=False) 

v.set(xlabel = "python", ylabel="seaborn")
sns.set(font_scale=2)
plt.show()
```
