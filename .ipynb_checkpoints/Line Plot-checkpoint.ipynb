# Line plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# plotting a basic line chart

axis_x = [1, 2, 3, 4, 5]
axis_y = [23, 43, 12, 45, 30]

sns.lineplot(x=axis_x, y=axis_y)

plt.show() # without this line, it shows extra useless things.
```

## import sample data -> Penguin

```python
# seaborn does its best when used with pandas!

# loading a csv data in the form of a dataframe from seaborn
data = sns.load_dataset('penguins')
data
```

```python
data = sns.load_dataset('penguins')
# creates labels for you automatically
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=data)
plt.show()
```

```python
# plotting data with respect to something

data = sns.load_dataset('penguins')
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=data, hue='sex')
plt.show()
```

```python
# adding size

data = sns.load_dataset('penguins')
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',size=50, data=data, hue='sex')
plt.show()
```

```python
# adding style

data = sns.load_dataset('penguins')
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data, hue='sex', style='sex') # separate line styles respective to sex
plt.show()
```

```python
# adding palette and markers

data = sns.load_dataset('penguins').head(40)
# palette only works when hue is present
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data, hue='sex', style='sex', palette="Accent", markers=['o', '>']) 
plt.show()
```

```python
# dash and removing legend

data = sns.load_dataset('penguins').head(40)
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data, hue='sex', style='sex', palette="Accent", markers=['o', '>'], dashes=False, legend=False) 
plt.show()
```

```python
# adding title and grid

data = sns.load_dataset('penguins').head(40)
sns.lineplot(x='bill_length_mm', y='bill_depth_mm',data=data, hue='sex', style='sex', palette="Accent", markers=['o', '>'], dashes=False, legend=False) 

plt.title("Penguins")
plt.grid()
plt.show()
```
